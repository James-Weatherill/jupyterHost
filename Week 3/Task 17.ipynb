```python
from math import trunc
from time import sleep

global curBal, curBalTruncRound
curBal = 105.50

def intro():
    print("Welcome to the Surrey Rock ATM!")
    menu()

def menu():
    usrInp = input("""\nPlease select an option below:
1: Display Balance
2: Withdraw Funds
3: Deposit Funds
9: Return Card

Enter here: """)
    
    if usrInp.isnumeric():
        if usrInp == "1":
            disBal()
        elif usrInp == "2":
            withFun()
        elif usrInp == "3":
            depFun()
        elif usrInp == "9":
            retCard()
        else:
            print("\nInvalid input, please select on of the options!")
            countdown = 3
            while countdown>0:
                print(countdown)
                countdown -= 1
                sleep(1)
            menu()
            
    else:
        print("\nInvalid input, please select on of the options!")
        countdown = 3
        while countdown>0:
            print(countdown)
            countdown -= 1
            sleep(1)
        menu()

def disBal():
    global curBal
    global curBalTruncRound
    
    print("\nYour current balance is: £" + str(curBal))
    curBalTrunc = trunc(curBal)
    curBalTruncRound = str(curBalTrunc)[:-1] + "0"
    print("\nThe maximum amount available for you to withdraw is: £" + curBalTruncRound)
    
    sleep(1)
    print("\n~\nYou will now return to the menu...\n~")
    sleep(2)
    menu()

def withFun():
    global curBal
    print("\nYour current balance is: £" + str(curBal))
    wthdrw = input("""\nPlease select an amount you would like to withdraw:
1: £10
2: £20
3: £30
4: £40
5: £60
6: £80
7: £100
8: Other Ammount
9: Return to Menu

Enter here: """)
    if wthdrw.isnumeric():
        if wthdrw == "1":
            if curBal >= 10:
                print("\nYou have withdrawn £10")
                curBal -= 10
                print("Your new balance is: £" + str(curBal))
                
                sleep(1)
                print("\n~\nYou will now return to the menu...\n~")
                sleep(2)
                menu()
            else:
                print("\nYou do not have sufficient funds")
                withFun()
        elif wthdrw == "2":
            if curBal >= 20:
                print("\nYou have withdrawn £20")
                curBal -= 20
                print("Your new balance is: £" + str(curBal))
                
                sleep(1)
                print("\n~\nYou will now return to the menu...\n~")
                sleep(2)
                menu()
            else:
                print("\nYou do not have sufficient funds")
                withFun()
        elif wthdrw == "3":
            if curBal >= 30:
                print("\nYou have withdrawn £30")
                curBal -= 30
                print("Your new balance is: £" + str(curBal))
                
                sleep(1)
                print("\n~\nYou will now return to the menu...\n~")
                sleep(2)
                menu()
            else:
                print("\nYou do not have sufficient funds")
                withFun()
        elif wthdrw == "4":
            if curBal >= 40:
                print("\nYou have withdrawn £40")
                curBal -= 40
                print("Your new balance is: £" + str(curBal))
                
                sleep(1)
                print("\n~\nYou will now return to the menu...\n~")
                sleep(2)
                menu()
            else:
                print("\nYou do not have sufficient funds")
                withFun()
        elif wthdrw == "5":
            if curBal >= 60:
                print("\nYou have withdrawn £60")
                curBal -= 60
                print("Your new balance is: £" + str(curBal))
                
                sleep(1)
                print("\n~\nYou will now return to the menu...\n~")
                sleep(2)
                menu()
            else:
                print("\nYou do not have sufficient funds")
                withFun()
        elif wthdrw == "6":
            if curBal >= 80:
                print("\nYou have withdrawn £80")
                curBal -= 80
                print("Your new balance is: £" + str(curBal))
                
                sleep(1)
                print("\n~\nYou will now return to the menu...\n~")
                sleep(2)
                menu()
            else:
                print("\nYou do not have sufficient funds")
                withFun()
        elif wthdrw == "7":
            if curBal >= 100:
                print("\nYou have withdrawn £100")
                curBal -= 100
                print("Your new balance is: £" + str(curBal))
                
                sleep(1)
                print("\n~\nYou will now return to the menu...\n~")
                sleep(2)
                menu()
            else:
                print("\nYou do not have sufficient funds")
                withFun()
        elif wthdrw == "8":
            ab = True
            while ab:
                try:
                    amount = float(input("\nPlease type a value you would like to withdraw: £"))
                    if curBal>=amount and (amount/(0.01))%1==0:
                        curBal -= amount
                        print("\nYou have withdrawn £" + str(amount) + "!")
                        print("\nYour new balance is: £" + str(curBal))
                        ab = False
                    else:
                        print("\nInvalid")
                        ab = False
                        
                        sleep(1)
                        print("\n~\nYou will now return to the menu...\n~")
                        sleep(2)
                        menu()
                        
                except:
                    print("\nInvalid input, try again!")
        elif wthdrw == "9":
            print("\n~\nYou will now return to the menu...\n~")
            sleep(2)
            menu()
        
    else:
        ("\nPlease try one of the numbered options!: ")
        countdown = 3
        while countdown>0:
            print(countdown)
            countdown -= 1
            sleep(1)
        withFun()
    
def depFun():
    global curBal
    
    depFunChoice = input("""\nPlease select an option from the list below:
1: Enter a deposit
2: Return to Menu

Enter here: """)
    if depFunChoice.isnumeric():
        if depFunChoice == "1":
            bc = True
            while bc:
                try:
                    amount = float(input("\nPlease type a value you would like to deposit: £"))
                    if (amount/(0.01))%1==0:
                        curBal += amount
                        print("\nYou have deposited £" + str(amount) + "!")
                        print("\nYour new balance is: £" + str(curBal))
                        bc = False
                        print("\n~\nYou will now return to the menu...\n~")
                        sleep(2)
                        menu()
                    else:
                        print("\nInvalid")
                        bc = False
                        
                        countdown = 3
                        while countdown>0:
                            print(countdown)
                            countdown -= 1
                            sleep(1)
                        depFun()
                except:
                    print("\nInvalid input, try again!")
        
        elif depFunChoice == "2":
            print("\n~\nYou will now return to the menu...\n~")
            sleep(2)
            menu()
        else:
            print("\nInvalid input, please choose an option from the menu!")
            countdown = 3
            while countdown>0:
                print(countdown)
                countdown -= 1
                sleep(1)
            depFun()        
    else:
        print("\nInvalid input, please choose an option from the menu!")
        countdown = 3
        while countdown>0:
            print(countdown)
            countdown -= 1
            sleep(1)
        depFun()  
        
def retCard():
    print("\nThank you for using Surrey Rock's ATM!\n\nGoodbye!")

intro()
```
