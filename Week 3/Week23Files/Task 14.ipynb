```python
# Pythagoras' Calculator

from math import sqrt
from time import sleep

def intro():
    print("The purpose of this program is to provide the\nanswer for any given side for a right angled triangle!")

    print("\nWe shall start with finding what you want to calculate...")

def menu():
    menuChoice = input("""\nPlease select an option from the Choices below

1: Find a Hypotenuse
2: Find an opposite / adjacent side
9: Exit the program

Enter Here: """)
    
    if menuChoice == "1":
        hypCalc()
    elif menuChoice == "2":
        sideCalc()
    elif menuChoice == "9":
        print ("\nGoodbye!")
    else:
        print("\nPlease choose one of the given options. Try again:\n")
        for i in range(1,4):
            print(4-i)
            sleep(1)
        menu()

def hypCalc():
    side1 = float(input("\nPlease type the length of a side: "))
    side2 = float(input("\nPlease type the length of the other side : "))
    
    hyp = sqrt((side1**2)+(side2**2))
    
    print("\nThe hypotenuse value for this triangle is:", hyp, "\n")
    for i in range(1,6):
        print(6-i)
        sleep(1)
    menu()

def sideCalc():
    hyp = float(input("\nPlease type the length of the Hypotenuse: "))
    side1 = float(input("\nPlease type the length of the short side: "))
    
    side2 = sqrt((hyp**2)-(side1**2))
    
    print("\nThe remaining value for this triangle is:", side2, "\n")
    for i in range(1,6):
        print(6-i)
        sleep(1)
    menu()

intro()
menu()
    
```
