```python
from math import trunc
from time import sleep

def main():
    notNum = True
    while notNum:
        usrNum1 = input("Please type an INTEGER: ")
        usrNum2 = input("Please type another INTEGER: ")

        if usrNum1.isnumeric() and usrNum2.isnumeric():
            usrNum1 = int(usrNum1)
            usrNum2 = int(usrNum2)
            notNum = False
        else:
            print("\nPlease make sure both numbers are INTEGERS\n")
            sleep(2)

    orderSwitch = True
    while orderSwitch:
        print("\nWe will now divide", str(usrNum1), "by", str(usrNum2) + ", and print the result and its remainder!")

        order = input("""\nIf you would like to switch the order of the numbers please type:
    'y': To Switch the numbers
    'n': To keep them as they are

    Enter here: """)
        order = order.lower()

        if order == 'y':
            temp = usrNum1
            usrNum1 = usrNum2
            usrNum2 = temp
        elif order == 'n':
            orderSwitch = False
        else:
            print("\n~~ " + str(order) + " is not one of the options. Try again... ~~")
            sleep(2)

    divided = (usrNum1/usrNum2)
    divtrunc = trunc(divided)

    remainderDeci = divided-divtrunc
    remainder = str(int(round((remainderDeci*usrNum2),1)))

    usrNum1 = str(usrNum1)
    usrNum2 = str(usrNum2)
    divtrunc = str(divtrunc)

    print("\n" + usrNum1 + " divided by " + usrNum2 + ", is: " + divtrunc + ", remainder " + remainder)
    sleep(2)
    
    restart()


def restart():
    restartQ = input("""\nWould you like to restart, with new numbers?
r: Restart
q: Quit

Enter here: """)
    restartQ = restartQ.lower()
    if restartQ == 'r':
        print("\n")
        main()
    elif restartQ == 'q':
        print("\nThank you, goodbye!")
    else:
        print("\nTry either 'r' or 'q'...")
        sleep(2)
        restart()

main()
```
