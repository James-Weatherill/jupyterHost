```python
from math import sqrt, trunc 
from time import sleep

def intro():
    print("""The purpose of this program is to find every prime number, between a given upper and lower bound!

The first thing will be to input a lower bound...""")
    main()

def main():
    primesList = []

    badInputs = True 
    while badInputs:

        lowerB = input("""\nPlease type a lower bound, it must be:
No less than 0 ~ 
No greater than 10,000 ~

Enter here: """)

        upperB = input("""\nNow, please type an upper bound, it must be:
No less than 0 ~ 
No greater than 10,000 ~ 
No more than 200 digits from the lower bound ~

Enter here: """)

        if lowerB.isnumeric() and upperB.isnumeric():
            lowerB = int(lowerB)
            upperB = int(upperB)
            if upperB-lowerB>200:
                print("\n\nMake sure the range is less than 201, try again!\n")
                sleep(2)
            elif upperB<lowerB:
                print("\n\nMake sure upper bound is greater than or equal to lower bound, try again!\n")
                sleep(2)
            else:
                if upperB>10000 or lowerB>10000:
                    print("\n\nMake sure inputs are less than 10,001, try again!\n")
                    sleep(2)
                else:
                    badInputs = False
        else:
            print("\n\nMake sure your inputs are positive integers, try again!\n")
            sleep(2)

    lowerConst = lowerB 
    add = 0 
    while lowerB < upperB: 
        lowerB = lowerConst + add 
        lowSqTr = trunc(sqrt(lowerB)) 
        for i in range(2,int(lowSqTr)+1): 
            if lowerB%i == 1: 
                prime = True 
            if lowerB%i == 0: 
                prime = False 
                break 
        if prime == True: 
            primesList.append(lowerB)
        add += 1

    print("\nThe primes between", lowerConst, "and", upperB, "are:\n" + str(primesList))
    sleep(2)
    print("\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")
    restart()

def restart():
    usrChoice = input("Would you like to Quit, or Restart and use different numbers?\n'q': Quit\n'r': Restart\n\nEnter here: ")
    usrChoice = usrChoice.lower()
    if usrChoice == 'q':
        print("\nGoodbye!")
    elif usrChoice == 'r':
        print("\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        main()
    else:
        print("\nInvalid input, try again with 'q' or 'r'!\n")
        restart()

intro()
```
