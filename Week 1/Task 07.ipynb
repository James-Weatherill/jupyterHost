```python
curNum1 = 0
totNum1 = 0

print('\n|Number|    |Total |\n|      |    |      |')
for i in range (1,101):
    
    curNum1 = i
    totNum1 = totNum1 + i
    
    print('|  ' + str(curNum1).zfill(3) + ' |    | ' + str(totNum1).zfill(4) + ' |')
```
