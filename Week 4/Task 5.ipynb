```python
from time import sleep

global FILEREAD
FILEREAD = open("colours.txt", "r+")

print("Hello, this program will show you the RGB codes for different colours!\nYou are also able to add your own colours!")

def homeMenu():
    global FILEREAD
    FILEREAD = open("colours.txt", "r+")
    
    read = readingFile()
    readRep = read.replace("\n"," ")
    readSplit = list(readRep.split(" "))
    
    usrStart = input("""\nWhat would you like to do?:
1: Find RGB codes
2: Add new colours
3: Quit

Enter here: """)
    
    if usrStart.isnumeric() and 0<int(usrStart)<=3:
        if usrStart == "1":
            main(read, readRep, readSplit)
        elif usrStart == "2":
            create(read, readRep, readSplit)
        else:
            print("\nThank you for using, goodbye!")
            FILEREAD.close()
    else:
        print("\nInvalid input! Please choose a number from the menu! ->")
        homeMenu()

def main(read, readRep, readSplit):
    global FILEREAD
    FILEREAD = open("colours.txt", "r+")
    
    usrLook = menuLookup(readSplit)
    
    for char in readSplit:
        if char == usrLook:
            loca = int(readSplit.index(char))
            print("\nID          = ", readSplit[loca])
            print("Colour      = ", readSplit[(loca+1)])
            print("Red Value   = ", readSplit[(loca+2)])
            print("Green Value = ", readSplit[(loca+3)])
            print("Blue Value  = ", readSplit[(loca+4)])
            
    print("\n\n~~ You will now return to the Main Menu! ~~\n")
    
    timer = 3
    while timer>(-1):
        print(timer)
        timer -= 1
        sleep(1)
        
    homeMenu()
                
def readingFile():
    global FILEREAD
    FILEREAD = open("colours.txt", "r+")
    
    fileR = FILEREAD.read()
    return fileR

def menuLookup(readSplit):
    global FILEREAD
    FILEREAD = open("colours.txt", "r+")
    
    badInps = True
    while badInps:
        print("\nPlease enter an ID for a colour you would like to display:\n")
        for num in readSplit:
            if (readSplit[(int(readSplit.index(num))+1)]).isnumeric() == False:
                numLoca = int(readSplit.index(num))
                print(readSplit[numLoca] + ":", readSplit[(numLoca+1)])
            
        usrInp = input("\nEnter here: ")
              
        if usrInp.isnumeric() and 0<int(usrInp)<=int(readSplit[numLoca]):
            badInps = False
            return usrInp 
        else:
            print("\nInvalid ID! Please choose an ID from the list above ^:")
            sleep(2)
    
def create(read, readRep, readSplit):
    global FILEREAD
    
    for num in readSplit:        
        if (readSplit[(int(readSplit.index(num))+1)]).isnumeric() == False:
            numLoca = readSplit.index(num)    
    
    colourName = input("\nPlease enter the name of the colour: ")
    colourName = colourName.lower()
    
    badInps = True
    while badInps:
        redValue = input("\nPlease enter the R (red) value: ")
        greenValue = input("\nPlease enter the G (green) value: ")
        blueValue = input("\nPlease enter the B (blue) value: ")
        if redValue.isnumeric() and greenValue.isnumeric() and blueValue.isnumeric and 0<=int(redValue)<=255 and 0<=int(greenValue)<=255 and 0<=int(blueValue)<=255:
            badInps = False
        else:
            print("\nAt least one of your inputs is incorrect! Try again ->")
            sleep(2)
    
    FILEREAD.write("\n")
    FILEREAD.write(str(int(readSplit[numLoca])+1) + " ")
    FILEREAD.write(colourName + " ")
    FILEREAD.write(redValue + " ")
    FILEREAD.write(greenValue + " ")
    FILEREAD.write(blueValue)
    
    FILEREAD.close()
    FILEREAD = open("colours.txt", "r+")
    
    print("\n\n~~ Thank you, you will now return to the Main Menu ~~\n")
    
    timer = 3
    while timer>(-1):
        print(timer)
        timer -= 1
        sleep(1)
        
    homeMenu()
    
homeMenu()
```
