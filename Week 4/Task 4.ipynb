```python
from time import sleep

print("Hello, this program will turn your mark into a grade!")

def usrInp():
    usrMark = input("\nPlease type the mark you recieved below!\n\nEnter here: ")
    testTotal = input("\nPlease now type the total marks of the entire test!\n\nEnter here: ")
    if usrMark.isnumeric() and testTotal.isnumeric():
        usrMark = int(usrMark)
        testTotal = int(testTotal)
        if usrMark<=testTotal:
            usrExpect(usrMark, testTotal)
        else:
            print("\nI don't think your mark is higher than the total! Try again ->\n")
            sleep(2)
            usrInp()
    else:
        print("\nPlease type a valid mark and total marks! Try again ->\n")
        sleep(2)
        usrInp()

def usrExpect(mark, total):
    usrGuess = input("""\nPlease choose from the menu below, what you think your grade will be ->
1: FAIL
2: Third
3: 2:2
4: 2:1
5: First

Enter here: """)
    if usrGuess == "1":
        mark_grade(mark, total, usrGuess)
    elif usrGuess == "2":
        mark_grade(mark, total, usrGuess)
    elif usrGuess == "3":
        mark_grade(mark, total, usrGuess)
    elif usrGuess == "4":
        mark_grade(mark, total, usrGuess)
    elif usrGuess == "5":
        mark_grade(mark, total, usrGuess)
    else:
        print("\nPlease enter one of the menu options! Try again ->\n")
        sleep(2)
        usrExpect()

def mark_grade(mark, total, guess):
    usrGradeDeci = mark/total
    if usrGradeDeci < 0.4:
        gradeCheck = "1"
        usrGrade = "FAIL"
    elif usrGradeDeci < 0.5:
        gradeCheck = "2"
        usrGrade = "Third"
    elif usrGradeDeci < 0.6:
        gradeCheck = "3"
        usrGrade = "2:2"
    elif usrGradeDeci < 0.7:
        gradeCheck = "4"
        usrGrade = "2:1"
    elif usrGradeDeci >= 0.7:
        gradeCheck = "5"
        usrGrade = "First"

    if guess == gradeCheck:
        print("\nYou guessed correctly! Your grade is a", usrGrade + "!")
    elif int(guess)>int(gradeCheck):
        print("\nUnfortunatley, you didn't perform as well as you hoped. Your grade is a", usrGrade +".")
    else:
        print("\nYou did even better than expected! The grade you acheived is a", usrGrade + "!")
    
usrInp()
```
